```python
# this step we will change all the doc into vec
# configure for input, modify this to fit u file
import numpy as np
input_file = "/home/qin/pockit/data/word2vec_input_jq.txt"
output_file = "/home/qin/pockit/data/word2vec_output.txt"
#========

import re
def clr(sent):
    return re.sub('[\r\n]','',sent)

def vec2json(result_dct, _lines,cnt):
    arrs= bc.encode(_lines)
    print('we have handled %s pcs'%cnt)
    for arr in arrs:
        cnt+=1
        result_dct[str(cnt)] = {}
        result_dct[str(cnt)]['sent'] = line
        result_dct[str(cnt)]['vec'] = ",".join([str(ar) for ar in arr])
    _lines = []
    return cnt


import bert_serving
from bert_serving.client import BertClient
bc = BertClient()
bc
arr = ''
cnt=0
result_dct = {}
with open(input_file,'r') as fin:
    lines = fin.readlines()
    #np.random.shuffle(lines)
    #lines = lines
    _lines = []
    for line in lines:
        line = clr(line)
        if line == "":
            continue
        _lines.append(line)
        if len(_lines)>0:
            if len(_lines)%500==0:
                cnt = vec2json(result_dct, _lines, cnt)
                _lines = []
    if len(_lines)>0:
        cnt = vec2json(result_dct, _lines, cnt)
        _lines = []

# note bar will  tell us how many sentences had be handler
```
